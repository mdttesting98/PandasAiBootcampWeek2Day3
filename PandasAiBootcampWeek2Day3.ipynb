````python
import numpy as np

dSum = np.sum(df['Production (short tons)'])
dMean = np.mean(df['Production (short tons)'])
dMax = np.max(df['Production (short tons)'])
dMin = np.min(df['Production (short tons)'])
               
print('Sum: ', dSum)
print('Mean: ', dMean)
print('Maximum: ', dMax)
print('Minimum: ', dMin)
````
